```python
col_list_outlier = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','DAYS_BIRTH']
for col in col_list_outlier: 
    q1 = application_data[col].quantile(0.25)
    q3 = application_data[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    application_data = application_data.loc[(application_data[col] > range_low) & (application_data[col] < range_high)]

application_data.shape
```

```python
#### Creating three age groups - Young, Mid age and Senior citizen
```

```python
def age_group(x):
    if (x < 30):
        return 'Young'
    elif (x >= 30 and x < 60):
        return 'Mid Age'
    else:
        return 'Senior'
```

```python
#Now Create a new column AGE_GROUP
application_data['AGE_GROUP'] = application_data['AGE'].apply(age_group)
application_data.head()
```
